# Indian population 

# importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Read the csv file

```python
df = pd.read_csv("state_wise_population__2019.csv")
```

```python
df.head()
```

# Check Columns

```python
df.columns
```

# 1) Bar Graph - Total Population by state

```python

plt.figure()
plt.bar(df['State'],df['total_population'])
plt.xlabel('State')
plt.ylabel('total_population')
plt.title('Total Population by state')
plt.xticks(rotation=90)
plt.show()
```

# 2) Pie chart - Male vs Female Population (overall)

```python
male_total = df['population_male'].sum()
female_total = df['population_female'].sum()

plt.figure()
plt.pie([male_total, female_total],
        labels=['Male', 'Female'],
        autopct='%1.1f%%')
plt.title('Male vs Female Population')
plt.show()
```

# 3) Line Graph - Population Trend Across States

```python
plt.figure()
plt.plot(df['State'], df['total_population'])
plt.xlabel('State')
plt.ylabel('Total Population')
plt.title('Line Graph of Population by State')
plt.xticks(rotation=90)
plt.show()
```

# 4) Scatter plot - Male vs Female Population

```python
plt.figure()
plt.scatter(df['population_male'], df['population_female'])
plt.xlabel('Male Population')
plt.ylabel('Female Population')
plt.title('Scatter Plot: Male vs Female Population')
plt.show()
```

# 5) Histogram - Distribution of Total Population

```python
plt.figure()
plt.hist(df['total_population'])
plt.xlabel('Total Population')
plt.ylabel('Frequency')
plt.title('Histogram of Total Population')
plt.show()
```

# 6) Area chart - Male & Female Population by state

```python
plt.figure()
plt.stackplot(df['State'],
              df['population_male'],
              df['population_female'],
              labels=['Male', 'Female'])
plt.xlabel('State')
plt.ylabel('Population')
plt.title('Area Chart of Male and Female Population')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
